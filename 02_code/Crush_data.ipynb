```python
file_path = '/Users/fengcharles/7023_group/01_data/02_raw/Motor_Vehicle_Collisions_-_Crashes_20240427.csv'

# Load the CSV file into a DataFrame
crash_data = pd.read_csv(file_path)
```

```python
crash_data
```

```python
columns_to_check = ['BOROUGH', 'ZIP CODE', 'LATITUDE','LONGITUDE','LOCATION']

# Drop rows where any of the specified columns have missing values
crash_data = crash_data.dropna(subset=columns_to_check)
```

```python
crash_data
```

```python
crash_data = gpd.GeoDataFrame(
    crash_data, 
    geometry=[Point(xy) for xy in zip(crash_data.LONGITUDE, crash_data.LATITUDE)]
)

# Set the coordinate reference system to EPSG:4269
crash_data.set_crs(epsg=4326, inplace=True,allow_override=True)
```

```python
output_path = '/Users/fengcharles/7023_group/01_data/03_intermediate/Crush_data.shp'

# Export the GeoDataFrame to a shapefile
crash_data.to_file(output_path, driver='ESRI Shapefile')
```
